```python
symbols_data = []
sec_name_data = []
sector_data = []
industry_data = []
close_price_data = []
previous_close_price_data = []
floating_shares_data = []
```

```python
 yf.Ticker("BF-B").info
```

## Replace dot in ticker with hypen for yahoo finance api to work correctly

```python
d = {'K': 1000, 'M': 1000000, 'B': 1000000000}

for ticker in sp500_tickers:
    try:
        modified_ticker = ticker.replace('-','.')
        print('Processing ticker ', ticker)
        tInfo = yf.Ticker(ticker).info

        ## Populate security info like name, ticker, sector and industry when applicable
        symbols_data.append(modified_ticker);
        sec_name_data.append(df[df['Symbol'] == modified_ticker]['Security'].values[0]);
        sector_data.append(df[df['Symbol'] == modified_ticker]['GICS Sector'].values[0]);

        if('industry' in tInfo):
            industry_data.append(tInfo['industry'])
        else:
            industry_data.append(None)

        ## Populate security price data
        quote = si.get_quote_table(ticker)

        close_price_data.append(quote['Quote Price'])
        previous_close_price_data.append(quote['Previous Close'])

        df1 = si.get_stats(ticker)
        floatShares = df1[df1['Attribute'] == 'Float']['Value'].values[0]
        floatshares_val = pd.to_numeric(floatShares[:-1]) * d[floatShares[-1]]
        floating_shares_data.append(floatshares_val)
    except:
        print('error processing ticker ', ticker)
```

### Populate all collected data into dataframe - TODO: Need to find a better way to do it

```python
out_data = {'Symbol': symbols_data,
            'Security': sec_name_data,
            'Sector': sector_data,
            'Industry': industry_data,
            'Close_Price': close_price_data,
            'Previous_Day_Close_Price': previous_close_price_data, 
            'Floating_Shares': floating_shares_data}
output_df = pd.DataFrame(out_data, columns= ['Symbol', 'Security', 'Sector', 'Industry','Close_Price','Previous_Day_Close_Price', 'Floating_Shares'])
```
