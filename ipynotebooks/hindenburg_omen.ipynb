```python
import pandas as pd
import yfinance as yf
import os
```

```python
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
constituents_table = payload[0]

df = constituents_table
high_low_percentage_threshold = 2;
```

```python
index_symbols = df['Symbol']
total_symbols = len(index_symbols)
```

```python
index_symbols.values
```

```python
symbols_data = []
sec_name_data = []
sector_data = []
high_52_wk_data = []
low_52_wk_data = []
today_closing_price_data = []
today_high_price_data = []
today_low_price_data = []
is_52_wk_high_data = []
is_52_wk_low_data = []
```

```python
for index, symbol in index_symbols.iteritems(): 
    if(symbol != 'BRK.B' and symbol != 'BF.B'):
        print('processing symbol', symbol)
        data = yf.Ticker(symbol).history(period="1y")
        sec_name_data.append(df[df['Symbol'] == symbol]['Security'].values[0]);
        sector_data.append(df[df['Symbol'] == symbol]['GICS Sector'].values[0]);
        high_52_wk = data['High'].max()
        low_52_wk = data['Low'].min()
        todays_high = data.head(1)['High'][0]
        todays_low = data.head(1)['Low'][0]
        todays_close = data.head(1)['Close'][0]

        symbols_data.append(symbol);
        high_52_wk_data.append(high_52_wk);
        low_52_wk_data.append(low_52_wk);
        today_closing_price_data.append(todays_close);
        today_high_price_data.append(todays_high);
        today_low_price_data.append(todays_low);

        is_52_wk_high = 'N';
        is_52_wk_low = 'N';


        if (todays_high >= (high_52_wk - ((high_low_percentage_threshold/100)* high_52_wk))):
            is_52_wk_high = 'Y';

        if (todays_high <= (low_52_wk + ((high_low_percentage_threshold/100)* low_52_wk))):
            is_52_wk_low = 'Y';

        is_52_wk_high_data.append(is_52_wk_high);
        is_52_wk_low_data.append(is_52_wk_low);
```

```python
out_data = {'Symbol': symbols_data,
            'Security': sec_name_data,
            'Sector': sector_data,
            '52_Wk_High': high_52_wk_data,
            '52_Wk_Low': low_52_wk_data, 
            'Todays_Close': today_closing_price_data, 
            'Todays_High': today_high_price_data, 
            'Todays_Low': today_low_price_data, 
            'Is_52_wk_high': is_52_wk_high_data,
            'Is_52_wk_Low': is_52_wk_low_data}
output_df = pd.DataFrame(out_data, columns= ['Symbol', 'Security', 'Sector', '52_Wk_High','52_Wk_Low','Todays_Close', 'Todays_High', 'Todays_Low', 'Is_52_wk_high', 'Is_52_wk_Low'])
```
