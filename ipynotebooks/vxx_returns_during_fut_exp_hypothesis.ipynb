```python
import pandas as pd
import os
import datetime
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
from datetime import timedelta
import yfinance as yf
```

```python
data_dir = '../'
data_file = 'vix_futures_historical_dates.csv'
file_path = os.path.join(data_dir, data_file)
```

```python
vix_fut_historical_exp_dates = pd.read_csv(file_path)
```

```python
vix_fut_historical_exp_dates['Fut_Exp_Date'] = pd.to_datetime(vix_fut_historical_exp_dates['Expiry_Date'], format='%m/%d/%Y')
vix_fut_historical_exp_dates['1_wk_Prior_Fut_Exp_Date'] = vix_fut_historical_exp_dates['Fut_Exp_Date'] - timedelta(days=7)
vix_fut_historical_exp_dates['1_wk_After_Fut_Exp_Date'] = vix_fut_historical_exp_dates['Fut_Exp_Date'] + timedelta(days=7)
```

```python
vxx_data = yf.Ticker('VXX').history(period="3y")
vix_data = yf.Ticker('^VIX').history(period="3y")
```

```python
vix_exp_dates = []
vix_price = []
vxx_price_on_exp = []
vxx_price_1_wk_bf_exp = []
vxx_price_1_wk_af_exp = []

for index, row in vix_fut_historical_exp_dates.iterrows():

    exp_date = row['Fut_Exp_Date'].strftime("%Y-%m-%d")
    one_wk_prior_dt = row['1_wk_Prior_Fut_Exp_Date'].strftime("%Y-%m-%d")
    one_wk_after_dt = row['1_wk_After_Fut_Exp_Date'].strftime("%Y-%m-%d")
    vix_exp_dates.append(exp_date)
    vix_price.append(vix_data.loc[exp_date]['Close'])
    vxx_price_on_exp.append(vxx_data.loc[exp_date]['Close']);
    vxx_price_1_wk_bf_exp.append(vxx_data.loc[one_wk_prior_dt]['Close']);
    vxx_price_1_wk_af_exp.append(vxx_data.loc[one_wk_after_dt]['Close']);
```

```python
out_data = {'Fut_Exp_Date': vix_exp_dates,
            'VIX': vix_price,
        'VXX_Price': vxx_price_on_exp, 
        'VXX_1_wk_Prior': vxx_price_1_wk_bf_exp,
        'VXX_1_wk_After': vxx_price_1_wk_af_exp}
output_df = pd.DataFrame(out_data, columns= ['Fut_Exp_Date', 'VIX','VXX_1_wk_Prior','VXX_Price', 'VXX_1_wk_After'])
```

```python
output_df['VXX_1_Wk_%_Returns_Before'] = ((output_df['VXX_Price'] - output_df['VXX_1_wk_Prior'])/output_df['VXX_1_wk_Prior']) * 100
output_df['VXX_1_Wk_%_Returns_After'] = ((output_df['VXX_1_wk_After'] - output_df['VXX_Price'])/output_df['VXX_Price']) * 100
```
