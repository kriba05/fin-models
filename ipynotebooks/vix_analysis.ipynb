```python
import pandas as pd
import os
import datetime
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
```

```python
symbol = "^VIX"
data_dir = '../data/%s' % symbol
data_file = '%s_history.csv' %symbol;
file_path = os.path.join(data_dir, data_file)
```

```python
data = pd.read_csv(file_path)
```

```python
data.head(5)
```

```python
analysis_price = 40
correction_price_pct = 0.5
correction_price = analysis_price - (analysis_price * correction_price_pct)
startDate = '2015-01-01'
endDate = '2021-02-28'
```

```python
analysis_data_window = data[(data['Date'] >= startDate) & (data['Date'] <= endDate)]
```

```python
analysis_data_window.head(5)
```

```python
analysis_data_window.tail(5)
```

```python
greater_than_analysis_price_data = analysis_data_window[analysis_data_window['Close'] > analysis_price]
```

```python
greater_than_analysis_price_data
```

```python
first_occurence = greater_than_analysis_price_data.iloc[0]
```

```python
first_occurence
```

```python
analysis_data_window.head(5)
```

```python
filtered_data_set = analysis_data_window[analysis_data_window['Date'] >= first_occurence['Date']]
```

```python
filtered_data_set.head(5)
```

```python
correction_price_date = filtered_data_set[filtered_data_set['Close'] < correction_price].iloc[0]['Date']
correction_price_date
```

```python
begin_date = datetime.datetime.strptime(first_occurence['Date'], '%Y-%m-%d')
end_date = datetime.datetime.strptime(correction_price_date, '%Y-%m-%d')
total_cal_days_for_correction =  end_date - begin_date
```

```python
total_calendar_days = total_cal_days_for_correction.days
total_calendar_days
```

```python
us_bd = CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

```python
num_of_business_days = len(pd.date_range(start = first_occurence['Date'], end = correction_price_date, freq = us_bd))
num_of_business_days
```

```python
print("Given Date range of {} and {} - Total number of business days taken for {} to drop from {} to {} is {} days (calendar_days_equivalent: {})".format(
      startDate, end_date, symbol, analysis_price, correction_price, num_of_business_days, total_calendar_days)
     )
```
